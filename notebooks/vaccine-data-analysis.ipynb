# Vaccination Data Analysis for Sweden Germany India

```python
#Importing necessary libraries
import pandas as pd
import subprocess
import os
import numpy as np
from datetime import datetime
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as pe
sns.set(style="whitegrid")
```

```python
#CSV value being read
vaccineData=r"C:\Users\adapa\OneDrive\Pictures\project\ads_covid_19_data_analysis\data\raw\vaccineData.csv"
vaccineData= pd.read_csv(vaccineData)
vaccineData.head()
```

# Data preparation

```python

dataOne= vaccineData.drop(['Doses_admin', 'People_partially_vaccinated', 'Report_Date_String', 'UID', 'Province_State'], axis=1, inplace=True)
dataTwo=vaccineData.rename(columns={'Country_Region': 'Country', 'People_fully_vaccinated': 'Vaccinated_People'})
dataTwo.head()
```

# Germany Sweden and India vaccine data

```python
dfGermany = dataTwo[dataTwo["Country"] == 'Germany'].copy()
dfGermany['Vaccinated_People'] = dfGermany['Vaccinated_People'].div(5531000)
dfSweden = dataTwo[dataTwo["Country"] == 'Sweden'].copy()
dfSweden['Vaccinated_People'] = dfSweden['Vaccinated_People'].div(771612)
dfIndia = dataTwo[dataTwo["Country"] == 'India'].copy()
dfIndia['Vaccinated_People'] = dfIndia['Vaccinated_People'].div(5831000)
```

```python
#NAN clearing
Germany = dfGermany.dropna()
Sweden = dfSweden.dropna()
India = dfIndia.dropna()
```

```python
#cList of chosen countries
dfCountry = [Germany, Sweden, India]
finalDf = pd.concat(dfCountry).reset_index(drop=True)
finalDf
```

```python
#Datformat change
finalDf['Date'] =  pd.to_datetime(finalDf['Date'], format='%Y-%m-%d')
```

```python
#Plot
plot = pe.line(finalDf, x="Date", y="Vaccinated_People", color = 'Country', title='Vaccination Rate Germany, Sweden, India')
plot.show()
```
